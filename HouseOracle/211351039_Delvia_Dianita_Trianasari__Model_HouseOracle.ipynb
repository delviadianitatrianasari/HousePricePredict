```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn.metrics import mean_absolute_percentage_error
```

```python
df = pd.read_csv("housesalesprediction/kc_house_data.csv")
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['id','date','sqft_basement','waterfront','sqft_above',
                 'zipcode', 'sqft_living15','sqft_lot15', 'view',
                 'lat', 'long', 'yr_renovated'], inplace=True)
df.head()
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Sequential()
model.add(Dense(32, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(32, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=32, epochs=50)
```

```python
y_pred = model.predict(X_test)
mape = mean_absolute_percentage_error(y_test, y_pred)
accuracy = 1 - mape
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
model.save('./houseprice.h5')
```

```python
model = tf.keras.models.load_model('./houseprice.h5')
converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_model = converter.convert()
open("./houseprice.tflite", "wb").write(tflite_model)
```
